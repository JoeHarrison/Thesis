# Increase Capacity while keeping lr

```python
def increase_capacity_keep_lr(network, capacity, optimizer, device):
    # Store old ids
    old_ids = [id(p) for p in network.parameters()]
    old_param_sizes = [p.size() for p in network.parameters()]

    network.increase_capacity(capacity)

    # Store new ids
    new_ids = [id(p) for p in network.parameters()]
    new_param_sizes = [p.size() for p in network.parameters()]

    # Store old state 
    opt_state_dict = optimizer.state_dict()
    for old_id, new_id, new_param_size, old_param_size in zip(old_ids, new_ids, new_param_sizes, old_param_sizes):
        # Store step, and exp_avgs
        step = opt_state_dict['state'][old_id]['step']
        old_exp_avg = opt_state_dict['state'][old_id]['exp_avg']
        old_exp_avg_sq = opt_state_dict['state'][old_id]['exp_avg_sq']
        old_max_exp_avg_sq = opt_state_dict['state'][old_id]['max_exp_avg_sq']

        exp_avg = torch.zeros(new_param_size)
        exp_avg_sq = torch.zeros(new_param_size)
        max_exp_avg_sq =  torch.zeros(new_param_size)
        # Extend exp_avgs to new shape depending on wether param is bias or weight
        if exp_avg.dim()>1:
            # Weights
            exp_avg[0:old_param_size[0],0:old_param_size[1]] = old_exp_avg
            exp_avg_sq[0:old_param_size[0],0:old_param_size[1]] = old_exp_avg_sq
            max_exp_avg_sq[0:old_param_size[0],0:old_param_size[1]] = old_max_exp_avg_sq
        else:
            # Biases/last layer
            exp_avg[0:old_param_size[0]] = old_exp_avg
            exp_avg_sq[0:old_param_size[0]] = old_exp_avg_sq
            max_exp_avg_sq[0:old_param_size[0]] = old_max_exp_avg_sq
        
        # Delete old id from state_dict and update new params and new id
        del opt_state_dict['state'][old_id]
        opt_state_dict['state'][new_id] = {
            'step': step,
            'exp_avg': exp_avg,
            'exp_avg_sq': exp_avg_sq.to(device),
            'max_exp_avg_sq' : max_exp_avg_sq.to(device)
        }
        opt_state_dict['param_groups'][0]['params'].remove(old_id)
        opt_state_dict['param_groups'][0]['params'].append(new_id)

    network.to(device)
    optimizer = optim.Adam(network.parameters(), amsgrad=True)
    optimizer.load_state_dict(opt_state_dict)
    
    return network, optimizer
```

# Data Generation

```python
def generate_zero_cartesian():
    return random.uniform(0, 49) / 100

def generate_one_cartesian():
    return random.uniform(50, 100) / 100

# def generate_both(num_data_points, p):
#     Xs, Ys = [], []
#     for _ in range(num_data_points*4):
#         if random.random() < p:
#             c = random.randint(0,3)
#             if c==0:
#                 # or(0, 0) -> 0
#                 Xs.append([0.0, 0.0, 0.0]); Ys.append([0])
#             elif c==1:
#                 # or(1, 0) -> 1
#                 Xs.append([1.0, 0.0, 0.0]); Ys.append([1])
#             elif c==2:
#                 # or(0, 1) -> 1
#                 Xs.append([0.0, 1.0, 0.0]); Ys.append([1])
#             else:
#                 # or(1, 1) -> 1
#                 Xs.append([1.0, 1.0, 0.0]); Ys.append([1])
#         else:
#             c = random.randint(0,3)
#             if c==0:
#                 # xor(0, 0) -> 0
#                 Xs.append([0.0, 0.0, 1.0]); Ys.append([0])
#             elif c==1:
#                 # xor(1, 0) -> 1
#                 Xs.append([1.0, 0.0, 1.0]); Ys.append([1])
#             elif c==2:
#                 # xor(0, 1) -> 1
#                 Xs.append([0.0, 1.0, 1.0]); Ys.append([1])
#             else:
#                 # xor(1, 1) -> 0
#                 Xs.append([1.0, 1.0, 1.0]); Ys.append([0])
#     Xs = torch.tensor(Xs, dtype=torch.float).to(device)
#     Ys = torch.tensor(Ys, dtype=torch.float).to(device)
#     return Xs, Ys

def generate_both(num_data_points, p):
    Xs, Ys = [], []
    for _ in range(num_data_points*4):
        if random.random() < p:
            c = random.randint(0,3)
            if c==0:
                # or(0, 0) -> 0
                Xs.append([generate_zero_cartesian(), generate_zero_cartesian(), 0.0]); Ys.append([0])
            elif c==1:
                # or(1, 0) -> 1
                Xs.append([generate_one_cartesian(), generate_zero_cartesian(), 0.0]); Ys.append([1])
            elif c==2:
                # or(0, 1) -> 1
                Xs.append([generate_zero_cartesian(), generate_one_cartesian(), 0.0]); Ys.append([1])
            else:
                # or(1, 1) -> 1
                Xs.append([generate_one_cartesian(), generate_one_cartesian(), 0.0]); Ys.append([1])
        else:
            c = random.randint(0,3)
            if c==0:
                # xor(0, 0) -> 0
                Xs.append([generate_zero_cartesian(), generate_zero_cartesian(), 1.0]); Ys.append([0])
            elif c==1:
                # xor(1, 0) -> 1
                Xs.append([generate_one_cartesian(), generate_zero_cartesian(), 1.0]); Ys.append([1])
            elif c==2:
                # xor(0, 1) -> 1
                Xs.append([generate_zero_cartesian(), generate_one_cartesian(), 1.0]); Ys.append([1])
            else:
                # xor(1, 1) -> 0
                Xs.append([generate_one_cartesian(), generate_one_cartesian(), 1.0]); Ys.append([0])
    Xs = torch.tensor(Xs, dtype=torch.float).to(device)
    Ys = torch.tensor(Ys, dtype=torch.float).to(device)
    return Xs, Ys 

# def generate_both(num_data_points, p):
#     Xs, Ys = [], []
#     for _ in range(num_data_points):
#         if random.random() < p:

#                 Xs.append([0.0, 0.0, 0.0]); Ys.append([0])

#                 Xs.append([1.0, 0.0, 0.0]); Ys.append([1])
  
#                 Xs.append([0.0, 1.0, 0.0]); Ys.append([1])

#                 Xs.append([1.0, 1.0, 0.0]); Ys.append([1])
#         else:


#                 Xs.append([0.0, 0.0, 1.0]); Ys.append([0])

#                 Xs.append([1.0, 0.0, 1.0]); Ys.append([1])

#                 Xs.append([0.0, 1.0, 1.0]); Ys.append([1])

#                 Xs.append([1.0, 1.0, 1.0]); Ys.append([0])
#     Xs = torch.tensor(Xs, dtype=torch.float).to(device)
#     Ys = torch.tensor(Ys, dtype=torch.float).to(device)
#     return Xs, Ys 
```

# Validation Set

```python
# Increase number to validate more points on the Cartesian plane
x = np.linspace(0, 1.0, 2)
y = np.linspace(0, 1.0, 2)

tmp_Xs_OR = []
tmp_Ys_OR = []
tmp_Xs_XOR = []
tmp_Ys_XOR = []
for i in y:
    for j in x:
        tmp_Xs_OR.append([i,j,0])
        tmp_Xs_XOR.append([i, j, 1])
        tmp_Ys_OR.append([int(i>0.5 or j>0.5)])
        tmp_Ys_XOR.append([int(i>0.5 and not j>0.5) or int(not i>0.5 and j>0.5)])

Xs_OR = torch.tensor(tmp_Xs_OR, dtype=torch.float).to(device)
Xs_XOR = torch.tensor(tmp_Xs_XOR, dtype=torch.float).to(device)
Ys_OR = torch.tensor(tmp_Ys_OR, dtype=torch.float).to(device)
Ys_XOR = torch.tensor(tmp_Ys_XOR, dtype=torch.float).to(device)
```

# XOR Experiments

```python
def experiment(initial_capacity, train_or, capacity, non_linearity, OR_p, XOR_p, threshold, keep_lr, iterations, seeds):
        t = time.time()
        OR_loss_before_training = []
        XOR_loss_before_training = []
        
        OR_loss_before_XOR_training = []
        XOR_loss_before_XOR_training = []
        
        losses_or = []
        losses_xor = []
        
        all_losses_OR = np.zeros(iterations*2)
        all_losses_XOR = np.zeros(iterations*2)
        
        for seed in range(seeds):
            # Set seeds
            random.seed(seed)
            np.random.seed(seed)
            torch.manual_seed(seed)
            
            # Initialisation network
            network = DQN(3, initial_capacity.copy(), 1, non_linearity).to(device)
            optimizer = optim.Adam(network.parameters(), amsgrad=True, lr=0.01)
            criterion = nn.BCELoss()
            
            XOR_loss = 0
            OR_loss = 0
            
            # Performance before training
            with torch.no_grad():
                prediction = non_linearity(network(Xs_OR))
                OR_loss = 1.0/(1.0+criterion(prediction, Ys_OR))
                OR_loss_before_training.append(OR_loss.item())

                prediction = non_linearity(network(Xs_XOR))
                XOR_loss = 1.0/(1.0+criterion(prediction, Ys_XOR))
                XOR_loss_before_training.append(XOR_loss.item())
            
            # Variable to keep track of loss plotting
            global_i = 0
            
            # OR training
            OR_iterations = 0
            if train_or:
                OR_iterations = iterations
                for i in range(OR_iterations):
                    optimizer.zero_grad()
                    Xs, Ys = generate_both(25, OR_p)
                    
                    prediction = non_linearity(network(Xs))
                    loss = criterion(prediction, Ys)
                    loss.backward()
                    optimizer.step()
                    
                    with torch.no_grad():
                        prediction = non_linearity(network(Xs_OR))
                        OR_loss = 1.0/(1.0+criterion(prediction, Ys_OR))
                        all_losses_OR[global_i] += OR_loss
                        
                        prediction = non_linearity(network(Xs_XOR))
                        XOR_loss = 1.0/(1.0+criterion(prediction, Ys_XOR))
                        all_losses_XOR[global_i] += XOR_loss
            
                    global_i += 1
                
                    # Fill rest of losses with last OR score
                    if OR_loss.item()>threshold:
                        
                        
                        for j in range(OR_iterations - global_i):
                            all_losses_OR[j + global_i] += OR_loss
                            all_losses_XOR[j + global_i] += XOR_loss
                        global_i += OR_iterations - global_i
                        break
            
            OR_loss_before_XOR_training.append(OR_loss.item())
            XOR_loss_before_XOR_training.append(XOR_loss.item())
            
            # Network capacity increase with or without keeping the learning rate
            if capacity is not None and train_or:
                if not keep_lr:
                    network.increase_capacity(capacity)
                    optimizer = optim.Adam(network.parameters(), amsgrad=True, lr=0.01)
                else:
                     network, optimizer = increase_capacity_keep_lr(network, capacity, optimizer, device)
                network.to(device)
            
            
            
            XOR_iterations = iterations
            if not train_or:
                XOR_iterations = iterations*2
                
            for i in range(XOR_iterations):
                if i == (iterations - 1) and not train_or and capacity is not None:
                    if not keep_lr:
                        network.increase_capacity(capacity)
                        optimizer = optim.Adam(network.parameters(), amsgrad=True, lr=0.01)
                    else:
                        network, optimizer = increase_capacity_keep_lr(network, capacity, optimizer, device)
                    network.to(device)
                    
                optimizer.zero_grad()
                Xs, Ys = generate_both(25, XOR_p)
                prediction = non_linearity(network(Xs))
                loss = criterion(prediction, Ys)
                loss.backward()
                optimizer.step()
                
                with torch.no_grad():
                    prediction = non_linearity(network(Xs_OR))
                    OR_loss = 1.0/(1.0+criterion(prediction, Ys_OR))
                    all_losses_OR[global_i] += OR_loss
                    
                    prediction = non_linearity(network(Xs_XOR))
                    XOR_loss = 1.0/(1.0+criterion(prediction, Ys_XOR))
                    all_losses_XOR[global_i] += XOR_loss
                
                global_i += 1
                
                if XOR_loss.item()>threshold:
                    for j in range(XOR_iterations + OR_iterations - global_i):
                        all_losses_OR[j + global_i] += OR_loss
                        all_losses_XOR[j + global_i] += XOR_loss
                    break
            
            average_loss = 0
            average_loss += OR_loss.item()
            losses_or.append(OR_loss.item())
            average_loss += XOR_loss.item()
            average_loss /= 2
            losses_xor.append(XOR_loss.item())
            
            
        # Print statistics
        print(initial_capacity, train_or, capacity, time.time()-t)
    
        print('Average loss or before training: ', np.average(OR_loss_before_training))
        print('Average loss xor before training: ', np.average(XOR_loss_before_training))
        print('Average loss before training: ', (np.average(OR_loss_before_training) + np.average(XOR_loss_before_training))/2)
        
        print('Average loss or before xor training: ', np.average(OR_loss_before_XOR_training))
        print('Average loss xor before xor training: ', np.average(XOR_loss_before_XOR_training))
        print('Average loss before xor training: ', (np.average(OR_loss_before_XOR_training) + np.average(XOR_loss_before_XOR_training))/2)
        
        print('Average loss or: ', np.average(losses_or))
        print('Average loss xor: ', np.average(losses_xor))
        print('Average loss: ', (np.average(losses_or) +  np.average(losses_xor))/2)
        
        plt.figure()
        if train_or:
            plt.axvline(x=OR_iterations, linestyle='dashed')
        plt.plot(all_losses_OR/seeds, label='OR Loss')
        plt.plot(all_losses_XOR/seeds, label='XOR Loss')
        plt.plot((all_losses_OR+ all_losses_XOR)/(2*seeds), label='Combined Loss')
        plt.xlabel('Steps')
        plt.ylabel('Loss')
        plt.legend()
        plt.show()
```

```python
non_linearity = torch.sigmoid
OR_p = 0.9
XOR_p = 0.1
threshold = 0.99
keep_lr = False
iterations = 1000
seeds = 5

experiment([1],False,None,non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([1], True, None,non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)

experiment([2],False,None,non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([1],False,[1],non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([2], True, None,non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([1], True, [1],non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)

experiment([3],False,None,non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([2], False, [1],non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([1], False, [2],non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([3], True, None,non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([2], True, [1],non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([1], True, [2],non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)

experiment([4],False,None,non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([3], False, [1],non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([2], False, [2],non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([1], False, [3],non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([4], True, None,non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([3], True, [1],non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([2], True, [2],non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([1], True, [3],non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)

experiment([5],False,None,non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([4], False, [1],non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([3], False, [2],non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([2], False, [3],non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([1], False, [4],non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([5], True, None,non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([4], True, [1],non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([3], True, [2],non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([2], True, [3],non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
experiment([1], True, [4],non_linearity,OR_p, XOR_p, threshold, keep_lr, iterations, seeds)
```
