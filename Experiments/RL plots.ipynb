```python
def maxDistance(arr): 
      
    # Used to store element to first index mapping 
    mp = {} 
  
    # Traverse elements and find maximum distance between 
    # same occurrences with the help of map. 
    maxDict = 0
    maxFirst_idx = 0
    for i in range(len(arr)): 
  
        # If this is first occurrence of element, insert its 
        # index in map 
        if arr[i] not in mp.keys(): 
            mp[arr[i]] = i 
  
        # Else update max distance 
        else: 
            if i-mp[arr[i]]>maxDict:
                maxDict = max(maxDict, i-mp[arr[i]])
                maxFirst_idx = mp[arr[i]]
  
    return maxDict, maxFirst_idx
```

```python
import torch
import rubiks2
from collections import defaultdict, deque
import time
from tqdm import tqdm_notebook

def n_scrambles(n, model):
    network = torch.load('./models/' + model + '/model.pt', map_location='cpu')
    network.to('cpu')
    device = 'cpu'
    
    solution_lengths = np.zeros(1000)
    solution_sequences = []
    
    for i in tqdm_notebook(range(1000)):
        env = rubiks2.RubiksEnv2(2, unsolved_reward=-1.0, seed=i)
        env.reset(n)
        hashes = defaultdict(list)
        done = 0
        tries = 0
        max_tries = 1000
        t = time.time()
        state = env.get_observation()
        hashes[hash(state.tostring())] = [0]*env.action_space.n
        stop = False
        state_hash_seq = []
        solution_sequence = []
        
        while time.time()-t<1.21 and not done and not stop:
            mask = hashes[hash(state.tostring())]
            state_hash_seq.append(hash(state.tostring()))
            action = network.act(state, 0.0, mask, device)
            solution_sequence.append(action)

            next_state, reward, done, info = env.step(action)  

            hstate = state.copy()
            state = next_state
            h = hash(state.tostring())
            if h in hashes.keys():      
                hashes[hash(hstate.tostring())][action] = -999
            else:
                hashes[h] = [0]*env.action_space.n

            tries += 1
            if max_tries:
                if tries >= max_tries:
                    stop = True

        length, first_idx = maxDistance(state_hash_seq)

        while length > 0:
            state_hash_seq = state_hash_seq[:first_idx] + state_hash_seq[first_idx + length:]
            solution_sequence = solution_sequence[:first_idx] + solution_sequence[first_idx + length:]
            length, first_idx = maxDistance(state_hash_seq)
            
        solution_lengths[i] = len(solution_sequence)
        solution_sequences.append(solution_sequence)
    
    return solution_lengths, solution_sequences
```

```python
naive_1000, _ = n_scrambles(1000, 'Naive_14')
lbf_1000, _ = n_scrambles(1000, 'LBF_14')

plt.figure()
plt.axhline(y=14, linestyle='dotted')
plt.boxplot([naive_1000, lbf_1000])
plt.title('1,000 Scrambles')
plt.ylabel('Number of Moves')
plt.xticks([1,2],['Naïve', 'LBF'])

naive_8, _ = n_scrambles(8, 'Naive_14')
lbf_8, _ = n_scrambles(8, 'LBF_14')

plt.figure()
plt.axhline(y=8, linestyle='dotted')
plt.boxplot([naive_8, lbf_8])
plt.title('8 Scrambles')
plt.ylabel('Number of Moves')
plt.xticks([1,2],['Naïve', 'LBF'])
```

```python
naive_1000, naive_1000_solution_seqs = n_scrambles(1000, 'Naive_14')
```

```python
triplets = defaultdict(int)

for solution_seq in naive_1000_solution_seqs:
    if len(solution_seq) >= 3:
        index_front = 0
        index_end = 3
        
        for i in range(len(solution_seq)-2):
            triplets[tuple(solution_seq[index_front:index_end])] += 1
            index_front+=1
            index_end+=1
    else:
        continue
        
aba = []
total_aba = 0
other = []
total_other = 0

for k,v in triplets.items():
    k = list(k)
    if tuple([k[0], k[2]]) in [(0,3),(1,4),(2,5),(3,0),(4,1),(5,2)]:
        aba.append(v)
        total_aba += v
    else:
        other.append(v)
        total_other += v

total = total_aba + total_other
m = max(max(aba), max(other))

aba_w = np.empty(np.array(aba).shape)
aba_w.fill(1/(np.array(aba).shape[0]+np.array(other).shape[0]))

other_w = np.empty(np.array(other).shape)
other_w.fill(1/(np.array(aba).shape[0]+np.array(other).shape[0]))

plt.figure()
plt.xlabel('Triplet Frequency')
plt.ylabel('Normalised Probability')
plt.hist(other, bins=np.arange(0, m), color='r', alpha=0.5, weights=other_w)
plt.hist(aba, bins=np.arange(0, m), color='b', alpha=0.5, weights=aba_w)

plt.legend(['Other', r'$aba^{-1}$'])
plt.axvline(np.array(other).mean(), color='r', linestyle='dashed', linewidth=1)
plt.axvline(np.array(aba).mean(), color='b', linestyle='dashed', linewidth=1)
plt.show()
```
