```python
def maxDistance(arr): 
      
    # Used to store element to first index mapping 
    mp = {} 
  
    # Traverse elements and find maximum distance between 
    # same occurrences with the help of map. 
    maxDict = 0
    maxFirst_idx = 0
    for i in range(len(arr)): 
  
        # If this is first occurrence of element, insert its 
        # index in map 
        if arr[i] not in mp.keys(): 
            mp[arr[i]] = i 
  
        # Else update max distance 
        else: 
            if i-mp[arr[i]]>maxDict:
                maxDict = max(maxDict, i-mp[arr[i]])
                maxFirst_idx = mp[arr[i]]
  
    return maxDict, maxFirst_idx
```

```python
import torch
import rubiks2
from collections import defaultdict, deque
import time
from tqdm import tqdm_notebook

def n_scrambles(n, model):
    network = torch.load('./models/' + model + '/model.pt', map_location='cpu')
    network.to('cuda')
    device = 'cuda'
    
    solution_lengths = np.zeros(1000)
    solution_sequences = []
    puzzles = []
    end_puzzles = []
    dones = 0
    
    for i in tqdm_notebook(range(2000)):
        env = rubiks2.RubiksEnv2(2, unsolved_reward=-1.0, seed=i*1000)
        env.reset(n)
        
        puzzles.append(env.get_observation())
        
        hashes = defaultdict(list)
        done = 0
        tries = 0
        t = time.time()
        
        max_tries = 1000
        
        state = env.get_observation()
        hashes[hash(state.tostring())] = [0]*env.action_space.n
        stop = False
        
        state_hash_seq = []
        solution_sequence = []
        
        while time.time()-t<1.21 and not done and not stop:
            mask = hashes[hash(state.tostring())]
            action = network.act(state, 0.0, mask, device)
            
            next_state, reward, done, info = env.step(action)
            
            state_hash_seq.append(hash(state.tostring()))
            solution_sequence.append(action) 

            hstate = state.copy()
            state = next_state
            h = hash(state.tostring())
            if h in hashes.keys():      
                hashes[hash(hstate.tostring())][action] = -999
            else:
                hashes[h] = [0]*env.action_space.n

            tries += 1
            if max_tries:
                if tries >= max_tries:
                    stop = True
        
        if not done:
            continue
            
        dones += done
        

        length, first_idx = maxDistance(state_hash_seq)

        while length > 0:
            state_hash_seq = state_hash_seq[:first_idx] + state_hash_seq[first_idx + length:]
            solution_sequence = solution_sequence[:first_idx] + solution_sequence[first_idx + length:]
            length, first_idx = maxDistance(state_hash_seq)
            
        solution_lengths[dones-1] = len(solution_sequence)
        solution_sequences.append(solution_sequence)
        
        env = rubiks2.RubiksEnv2(2, unsolved_reward=-1.0, seed=i)
        env.reset(n)
        for action in solution_sequence:
            env.step(int(action))
        end_puzzles.append(env.get_observation())
        
        if dones==1000:
            break
    
    return solution_lengths, solution_sequences, puzzles, end_puzzles, dones

```

```python
naive_1000, naive_1000_solution_seqs, _, _, naive_dones = n_scrambles(1000, 'Naive_14')
lbf_1000, lbf_1000_solution_seqs, _, _, lbf_dones = n_scrambles(1000, 'LBF_14')
deep_naive_1000, deep_naive_1000_solution_seqs, _, _, deep_naive_dones = n_scrambles(1000, 'Deepneat_14_naive')
deep_lbf_1000, deep_lbf_1000_solution_seqs, _, _, deep_lbf_dones = n_scrambles(1000, 'Deepneat_14_LBF')
print(naive_dones/1000.0, lbf_dones/1000.0, deep_naive_dones/1000.0, deep_lbf_dones/1000.0)

naive_8, _, _, _, _ = n_scrambles(8, 'Naive_14')
lbf_8, _, _, _, _ = n_scrambles(8, 'LBF_14')
deep_naive_8, _, _, _, _ = n_scrambles(8, 'Deepneat_14_naive')
deep_lbf_8, _, _, _, _ = n_scrambles(8, 'Deepneat_14_LBF')
```
